# Image to Pencil Sketch using OpenCV

# Load Libraries

```python
import cv2
from google.colab.patches import cv2_imshow

import requests
```

# Load Image

```python
url = 'https://media.istockphoto.com/id/1146517111/photo/taj-mahal-mausoleum-in-agra.jpg?s=612x612&w=0&k=20&c=vcIjhwUrNyjoKbGbAQ5sOcEzDUgOfCsm9ySmJ8gNeRk='
response = requests.get(url)
with open("image.jpg", "wb") as f:
    f.write(response.content)
```

# Transform

```python
grey_img = cv2.imread('/content/image.jpg', 0)  # gray_scale
invert = cv2.bitwise_not(grey_img)  # inverted -- negative img
# sharp edges in images are smoothed while minimizing too much blurring
blur = cv2.GaussianBlur(invert, (21, 21), 0)
invertedblur = cv2.bitwise_not(blur)
```

```python
sketch = cv2.divide(grey_img, invertedblur, scale=256.0)
cv2_imshow(sketch)
```
